### The updated plot

```python
# lets try and make this more palatable
methods = [
        "missForest", "Zero", "Mean",
        "Low value", "Sample from Gaussian dist.",
        "Double Boundary Bayes", "kNN", "minProb",
        "Random forest", "Other"]

counts = [4, 2, 1, 15, 28, 2, 6, 2, 2, 4]

plotter_update = pd.DataFrame(columns=['method', 'counts'])
plotter_update.method = methods
plotter_update.counts = counts

# sort
plotter_update = plotter_update.sort_values(by="counts")

# plot
plt.figure(figsize=(6, 5))

ax = sns.barplot(x="counts", y="method", data=plotter_update)
ax.set_xlabel('citation count')
ax.set_ylabel('')

#plt.show()
#plt.tight_layout()
#plt.savefig('imp-lit-seq-results-update.png', dpi=250, bbox_inches="tight")
```
